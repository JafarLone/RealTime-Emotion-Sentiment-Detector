```python
%%writefile my_project/app.py

import streamlit as st
import cv2
import tempfile
import base64

# Function to set a brighter background image
def add_bg_from_local(image_file):
    with open(image_file, "rb") as file:
        encoded_image = base64.b64encode(file.read()).decode()
    st.markdown(
        f"""
        <style>
        .stApp {{
            background-image: url("data:image/png;base64,{encoded_image}");
            background-size: cover;
            background-repeat: no-repeat;
            background-attachment: fixed;
            background-position: center;
            filter: brightness(0.8);  /* Lighter overlay for a brighter effect */
        }}
        </style>
        """,
        unsafe_allow_html=True
    )

# Apply the brightened background image
add_bg_from_local('/content/my_project/assets/background.png')

# Custom CSS for styling the header, sidebar, and other elements
st.markdown("""
    <style>
    /* Header Styling */
    .css-18e3th9 {
        text-align: center;
        color: #f0f0f0;  /* Lighter text for the header */
        font-size: 3em;
        font-weight: bold;
        margin-top: -50px;
        text-shadow: 2px 2px 8px rgba(0, 0, 0, 0.3);  /* Slight shadow for text */
    }

    /* Sidebar styling */
    .css-1d391kg {
        background-color: rgba(255, 255, 255, 0.2); /* Lightened background */
        border: 1px solid #f0f0f0; /* Lighter border */
        border-radius: 10px;
        padding: 20px;
        color: #ffffff;
        box-shadow: 0px 0px 10px rgba(0, 0, 0, 0.1); /* Light shadow for sidebar */
    }

    /* File uploader and buttons */
    .stButton button {
        background-color: #4CAF50;  /* Bright green button */
        color: white;
        padding: 12px 24px;
        border-radius: 8px;
        font-size: 1.2em;
        box-shadow: 0px 4px 8px rgba(0, 0, 0, 0.2);
    }

    /* Video container */
    .stVideo {
        display: flex;
        justify-content: center;
        align-items: center;
        padding: 20px;
    }

    /* General adjustments */
    .stApp {
        padding-top: 50px;
    }
    </style>
""", unsafe_allow_html=True)
# Title of the app (centered and styled)
# Title of the app (centered and styled)
# Title of the app (centered and styled at the top of the page)
st.markdown("""
    <style>
        .stApp {
            padding-top: 0px; /* Remove padding at the top */
        }
    </style>
    <h1 style="text-align:center; color:black; margin-top:0;">
        Emotion & Sentiment AI
    </h1>
""", unsafe_allow_html=True)



# Sidebar for Upload and Camera Capture options
st.sidebar.title("Upload or Capture Video 🎞️")
option = st.sidebar.radio("Choose an option:", ("Upload Video", "Capture Live Stream"))

# Limit the video size to a reasonable size (200MB in this case)
max_file_size_mb = 200
st.sidebar.title(f"Max file size: {max_file_size_mb} MB")
video_file = st.sidebar.file_uploader("Upload a video", type=["mp4", "avi", "mov"])

# Handling video upload
if option == "Upload Video" and video_file:
    # Check file size before processing
    file_size_mb = len(video_file.getvalue()) / (1024 * 1024)  # Convert bytes to MB
    if file_size_mb > max_file_size_mb:
        st.error(f"File is too large. Max file size allowed is {max_file_size_mb} MB.")
    else:
        st.success(f"Video uploaded successfully! Size: {file_size_mb:.2f} MB")

        # Save the uploaded video to a temporary location
        tfile = tempfile.NamedTemporaryFile(delete=False)
        tfile.write(video_file.read())

        # Display the uploaded video
        st.video(tfile.name)  # To confirm the video file is uploaded and saved correctly

        # Open the video file using OpenCV
        cap = cv2.VideoCapture(tfile.name)

        stframe = st.empty()  # Create a placeholder for video frames

        # Loop through the video frames and display them
        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                st.warning("End of video or cannot read the file.")
                break

            # Resize the frame for better viewing experience
            frame = cv2.resize(frame, (640, 480))

            # Display the video frames in the placeholder
            stframe.image(frame, channels="BGR")

        cap.release()

elif option == "Capture Live Stream":
    st.write("Capturing live stream...")
    st.warning("Live capture is not supported in Colab.")
```
